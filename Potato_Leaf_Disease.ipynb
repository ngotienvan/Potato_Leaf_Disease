```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file results.csv
df = pd.read_csv('/content/drive/MyDrive/doanchuyennganhai/doan2/yolo_runs_potato/potato_leaf_disease_train/results.csv')
df.columns = df.columns.str.strip()

# Tạo biểu đồ
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(14, 10))

# Train loss
sns.lineplot(x='epoch', y='train/box_loss', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Train Box Loss')

sns.lineplot(x='epoch', y='train/cls_loss', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Train Class Loss')

# Validation loss
sns.lineplot(x='epoch', y='val/box_loss', data=df, ax=axs[1, 0])
axs[1, 0].set_title('Validation Box Loss')

sns.lineplot(x='epoch', y='val/cls_loss', data=df, ax=axs[1, 1])
axs[1, 1].set_title('Validation Class Loss')

# DFL loss (distribution focal loss)
sns.lineplot(x='epoch', y='train/dfl_loss', data=df, ax=axs[2, 0])
axs[2, 0].set_title('Train DFL Loss')

sns.lineplot(x='epoch', y='val/dfl_loss', data=df, ax=axs[2, 1])
axs[2, 1].set_title('Validation DFL Loss')

plt.tight_layout()
plt.suptitle("📈 Diễn biến Loss trong quá trình huấn luyện", fontsize=18, y=1.02)
plt.show()
```

```python
%%writefile app.py
import streamlit as st
from PIL import Image, ImageEnhance, ImageDraw
import numpy as np
from ultralytics import YOLO
import logging

# Ghi log lỗi vào file
logging.basicConfig(filename='app.log', level=logging.DEBUG,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# Load mô hình YOLO đã huấn luyện
model = YOLO("/content/drive/MyDrive/doanchuyennganhai/doan2/yolo_runs_potato/potato_leaf_disease_train/weights/best.pt")

# Thiết lập giao diện trang
st.set_page_config(page_title="Nhận diện và phân loại bệnh lá khoai tây", layout="centered")

# ==== Giao diện CSS theo phong cách thiên nhiên ====
custom_css = """
<style>
body, .stApp {
    background-color: #e9f5e1;  /* Xanh lá nhạt */
    color: #1b4332;             /* Xanh đậm */
    font-family: 'Segoe UI', sans-serif;
}
h1, h2, h3, h4 {
    color: #2d6a4f;
}
label, .stSlider label, .stMarkdown {
    color: #2d6a4f !important;
}
.stButton>button {
    background-color: #52b788;
    color: white;
    border-radius: 10px;
    font-weight: bold;
}
.stButton>button:hover {
    background-color: #40916c;
}
.stFileUploader label {
    color: #2d6a4f;
    font-weight: bold;
}
</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)

# ==== Giao diện chính ====
st.title("🌿 Hệ thống phân loại tình trạng bệnh lá khoai tây")
uploaded_file = st.file_uploader("📤 Tải ảnh lá khoai tây lên", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    try:
        image = Image.open(uploaded_file).convert("RGB")

        # Điều chỉnh độ sáng ảnh
        brightness = st.slider("🌞 Điều chỉnh độ sáng ảnh", min_value=0.5, max_value=2.0, value=1.0, step=0.1)
        enhancer = ImageEnhance.Brightness(image)
        bright_image = enhancer.enhance(brightness)

        # Dự đoán bằng mô hình
        results = model.predict(bright_image)

        # Xử lý kết quả và vẽ bounding box
        processed_image = bright_image.copy()
        draw = ImageDraw.Draw(processed_image)

        detected_label = "Không xác định"
        highest_conf = 0

        if results and results[0].boxes is not None:
            result = results[0]
            for box in result.boxes:
                x1, y1, x2, y2 = box.xyxy[0].tolist()
                conf = float(box.conf[0])
                cls = int(box.cls[0])
                label = model.names[cls]

                if conf > highest_conf:
                    detected_label = label
                    highest_conf = conf

                draw.rectangle([x1, y1, x2, y2], outline="green", width=3)
                draw.text((x1, y1 - 10), f"{label} ({conf:.2f})", fill="green")

        # Hiển thị ảnh gốc và ảnh đã xử lý
        col1, col2 = st.columns(2)
        with col1:
            st.image(bright_image, caption="📷 Ảnh gốc đã chỉnh sáng", use_container_width=True)
        with col2:
            st.image(processed_image, caption="🧪 Ảnh sau khi nhận diện", use_container_width=True)

        # Hiển thị kết quả nhận diện
        st.markdown(f"<div style='color: black; font-weight: bold; font-size: 18px;'>✅ Kết quả: {detected_label}</div>", unsafe_allow_html=True)

        # Gợi ý xử lý dựa trên loại bệnh
        advice = ""
        if detected_label == "Potato_Early_blight":
            advice = (
                "👉 Cách xử lý \n"
                "- Loại bỏ lá bị bệnh để hạn chế lây lan.\n"
                "- Dùng thuốc có hoạt chất Mancozeb hoặc Chlorothalonil.\n"
                "- Giữ vườn thông thoáng, tránh độ ẩm cao."
            )
        elif detected_label == "Potato_Late_blight":
            advice = (
                "👉 Cách xử lý \n"
                "- Cắt và tiêu hủy cây bị nặng.\n"
                "- Phun thuốc Metalaxyl hoặc Fosetyl-Al.\n"
                "- Không tưới lên lá vào chiều tối."
            )
        elif detected_label == "Potato_healthy":
            advice = (
                "🌿 **Lá khoai tây khỏe mạnh!**\n"
                "- Tiếp tục chăm sóc như hiện tại.\n"
                "- Kiểm tra định kỳ để phát hiện bệnh sớm.\n"
                "- Có thể dùng chế phẩm sinh học để tăng đề kháng."
            )
        else:
            advice = (
                "⚠️ Không xác định được loại bệnh cụ thể.\n"
                "- Hãy thử lại với ảnh rõ hơn hoặc từ góc chụp khác."
            )

        # Hiển thị gợi ý
        st.markdown(f"<div style='background-color: #f0fdf4; padding: 15px; border-radius: 10px;'><b>{advice}</b></div>", unsafe_allow_html=True)

    except Exception as e:
        logging.exception(f"Lỗi xử lý ảnh: {str(e)}")
        st.error(f"❌ Đã xảy ra lỗi: {str(e)}")
```
